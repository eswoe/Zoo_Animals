```python
import pandas as pd
```

```python
#RSQ:
#What taxonomic class of animal survives longest in captivity? Which survives the shortest amount of time?
```

```python
animals_df = pd.read_csv('Downloads/animals_dataset.csv', delimiter=',')
```

```python
pd.options.display.max_rows = 100
```

```python
animals_df = animals_df.set_index('Species Common Name')
```

```python
animals_df
```

```python
animals_df = animals_df.rename(columns={'Overall MLE': 'Overall Median Life Expectancy'})
animals_df = animals_df.rename(columns={'Male MLE': 'Male Median Life Expectancy'})
animals_df = animals_df.rename(columns={'Female MLE': 'Female Median Life Expectancy'})
animals_df = animals_df.drop(columns='')
animals_df
```

```python
animals_df.describe(include='all')
```

```python
#Insights:
    #There are 6 unique taxonomic classes of animals represesented in the dataset
    #The taxonomic class with the greatest representation is Mammalia with a count of 175 out of 330
    #The average Overall Median Life Expectancy for all animals in captivity is 13.40 years
    #The minimum Overall Median Life Expectancy for all animals in captivity is 2.40 years
    #The maximum Overall Median Life Expectancy for all animals in captivity is 47.00 years
    
    #The average Male Median Life Expectancy for animals in captivity is 13.39 years
    #The average Female Median Life Expectancy for animals in captivity is 13.45 years
```
