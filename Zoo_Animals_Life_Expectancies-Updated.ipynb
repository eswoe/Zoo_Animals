```python
#RSQ:
#What taxonomic class of animal survives longest in captivity? Which survives the shortest amount of time? How do
#male and female median life expectancies compare within the same species?
```

```python
animals_df = pd.read_csv('animals_dataset.csv', delimiter=',')
```

```python
pd.options.display.max_rows = 100
```

```python
animals_df["ID"] = animals_df.index + 1
animals_df.set_index("ID")
```
